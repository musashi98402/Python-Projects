```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from plotnine.data import economics 
economics
econ = pd.DataFrame(economics)
econ = econ.set_index('date')
plt.plot(econ['unemploy'])
```

```python


fig, ax = plt.subplots(figsize = (8,5), dpi=125, facecolor='#12273f')
# deal with spines
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_color('#c4c9ce')
ax.spines['bottom'].set_linewidth(4)
ax.xaxis.set_tick_params(width=3)
ax.yaxis.set_tick_params(width=0)
# add grid lines
ax.set_axisbelow(True)
ax.yaxis.grid(color='#c4c9ce', linestyle='-')
# flip label to right spine 
ax.yaxis.set_label_position("right")
ax.yaxis.tick_right()
# include origin point in y axis
ax.set_ylim(0,econ['unemploy'].max()+750)
# set axes' and ticks' and their colors to grey 
ax.tick_params(axis='x', colors='#c4c9ce')
ax.tick_params(axis='y', colors='#c4c9ce')
ax.set_yticks(np.arange(0, max(econ['unemploy']), 2500))
# increase fontsizes
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
# set face colour
ax.set_facecolor("#12273f")
# plot the time series
ax.plot(econ['unemploy'], color = '#d4af37', label = 'Unemployment')
# ax.legend(loc='best', fancybox=True, framealpha=0, labelcolor='white', fontsize=18)
# adding inline label 
plt.annotate('Unemployment', xy=(econ.index.mean(), int(econ['unemploy'].mean()+2500)), xycoords='data', color='#d4af37')
# adding title 
plt.title('Unemployment Over Time',  color='white', fontname='arial', fontsize=24)
plt.show()
```
