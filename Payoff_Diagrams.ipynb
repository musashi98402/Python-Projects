```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
# Bull and Bear Spreads

S = np.arange(100,251,1) # underlying asset

# Calls
Kc1 = 150
Kc2 = 170
Nc1 = 1
Nc2 = 1
Pc1 = 20
Pc2 = 15

# Puts
Kp1 = 170
Kp2 = 150
Np1 = 1
Np2 = 1
Pp1 = 20
Pp2 = 15


long_call = [Nc1*max(-Pc1, x-Kc1-Pc1 ) for x in S]
short_call = [Nc2*min(Pc2, -(x-Kc2-Pc2)) for x in S]
bull_spread = np.sum([long_call, short_call], axis=0)

long_put = [Np1*max(-Pp1, Kp1-x-Pp1 ) for x in S]
short_put = [Np2*min(Pp2, -(Kp2-x-Pp2)) for x in S]
bear_spread = np.sum([long_put, short_put], axis=0)



plt.figure(figsize=(8,5))
ax = plt.gca()
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.plot(S, long_call, linestyle = '--')
plt.plot(S, short_call, linestyle = '--')
plt.plot(S, bull_spread)
plt.fill_between(S,bull_spread,0, where=bull_spread<0, color='r', alpha=0.1, label = 'Loss')
plt.fill_between(S,bull_spread,0, where=bull_spread>0, color='g', alpha=0.1, label = 'Profit')
plt.title('Bull Spread');


plt.figure(figsize=(8,5))
ax = plt.gca()
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.plot(S, long_put, linestyle = '--')
plt.plot(S, short_put, linestyle = '--')
plt.plot(S, bear_spread)
plt.fill_between(S,bear_spread,0, where=bear_spread<0, color='r', alpha=0.1, label = 'Loss')
plt.fill_between(S,bear_spread,0, where=bear_spread>0, color='g', alpha=0.1, label = 'Profit')
plt.title('Bear Spread');

```

```python
# Straddles

S = np.arange(100,251,1) 
# Longs
Kc1 = 175
Nc1 = 1
Pc1 = 10
Kp1 = 175
Np1 = 1
Pp1 = 10


# Shorts
Kc2 = 175
Nc2 = 1
Pc2 = 10
Kp2 = 175
Np2 = 1
Pp2 = 10

long_call = [Nc1*max(-Pc1, x-Kc1-Pc1 ) for x in S]
long_put = [Np1*max(-Pp1, Kp1-x-Pp1 ) for x in S]
long_straddle = np.sum([long_put, long_call], axis=0)

short_call = [Nc2*min(Pc2, -(x-Kc2-Pc2)) for x in S]
short_put = [Np2*min(Pp2, -(Kp2-x-Pp2)) for x in S]
short_straddle = np.sum([short_put, short_call], axis=0)


plt.figure(figsize=(8,5))
ax = plt.gca()
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.plot(S, long_put, linestyle = '--')
plt.plot(S, long_call, linestyle = '--')
plt.plot(S, long_straddle)
plt.fill_between(S,long_straddle,0, where=long_straddle<0, color='r', alpha=0.1, label = 'Loss')
plt.fill_between(S,long_straddle,0, where=long_straddle>0, color='g', alpha=0.1, label = 'Profit')
plt.title('Long Straddle');


plt.figure(figsize=(8,5))
ax = plt.gca()
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.plot(S, short_put, linestyle = '--')
plt.plot(S, short_call, linestyle = '--')
plt.plot(S, short_straddle)
plt.fill_between(S,short_straddle,0, where=short_straddle<0, color='r', alpha=0.1, label = 'Loss')
plt.fill_between(S,short_straddle,0, where=short_straddle>0, color='g', alpha=0.1, label = 'Profit')
plt.title('Short Straddle');
```

```python
# Strangles
S = np.arange(100,251,1) 
# Longs
Kc1 = 185
Nc1 = 1
Pc1 = 10
Kp1 = 165
Np1 = 1
Pp1 = 10

long_call = [Nc1*max(-Pc1, x-Kc1-Pc1 ) for x in S]
long_put = [Np1*max(-Pp1, Kp1-x-Pp1 ) for x in S]
long_strangle = np.sum([long_put, long_call], axis=0)


# Shorts
Kc2 = 185
Nc2 = 1
Pc2 = 10
Kp2 = 165
Np2 = 1
Pp2 = 10

short_call = [Nc2*min(Pc2, -(x-Kc2-Pc2)) for x in S]
short_put = [Np2*min(Pp2, -(Kp2-x-Pp2)) for x in S]
short_strangle = np.sum([short_put, short_call], axis=0)


plt.figure(figsize=(8,5))
ax = plt.gca()
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.plot(S, long_put, linestyle = '--')
plt.plot(S, long_call, linestyle = '--')
plt.fill_between(S,long_strangle,0, where=long_strangle<0, color='r', alpha=0.1, label = 'Loss')
plt.fill_between(S,long_strangle,0, where=long_strangle>0, color='g', alpha=0.1, label = 'Profit')
plt.plot(S, long_strangle)


plt.figure(figsize=(8,5))
ax = plt.gca()
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.plot(S, short_put, linestyle = '--')
plt.plot(S, short_call, linestyle = '--')
plt.plot(S, short_strangle)
plt.fill_between(S,short_strangle,0, where=short_strangle<0, color='r', alpha=0.1, label = 'Loss')
plt.fill_between(S,short_strangle,0, where=short_strangle>0, color='g', alpha=0.1, label = 'Profit')
plt.title('Short Strangle');
```

```python
# Butterfly Spreads

# Calls:
# ITM
Kc1 = 165
Nc1 = 1
Pc1 = 30
# OTM
Kc2 = 185
Nc2 = 1
Pc2 = 10
# ATM
Kc3 = 175
Nc3 = 2
Pc3 = 20

# Long Butterfly:


long_call_1 = [Nc1*max(-Pc1, x-Kc1-Pc1 ) for x in S]
long_call_2 = [Nc2*max(-Pc2, x-Kc2-Pc2 ) for x in S]
short_calls = [Nc3*min(Pc3, -(x-Kc3-Pc3)) for x in S]
long_butterfly = np.sum([long_call_1, long_call_2, short_calls], axis=0)


# Short Butterfly:
short_call_1 = [Nc1*min(Pc1, -(x-Kc1-Pc1)) for x in S]
short_call_2 = [Nc2*min(Pc2, -(x-Kc2-Pc2)) for x in S]
long_calls = [Nc3*max(-Pc3, x-Kc3-Pc3 ) for x in S]
short_butterfly = np.sum([short_call_1, short_call_2, long_calls], axis=0)


plt.figure(figsize=(8,5))
ax = plt.gca()
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.plot(S, short_calls, linestyle = '--')
plt.plot(S, long_call_1, linestyle = '--')
plt.plot(S, long_call_2, linestyle = '--')
plt.title('Long Butterfly Spread')
plt.plot(S, long_butterfly)
plt.fill_between(S,long_butterfly,0, where=long_butterfly<0, color='r', alpha=0.1, label = 'Loss')
plt.fill_between(S,long_butterfly,0, where=long_butterfly>0, color='g', alpha=0.1, label = 'Profit');

plt.figure(figsize=(8,5))
ax = plt.gca()
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.plot(S, long_calls, linestyle = '--')
plt.plot(S, short_call_1, linestyle = '--')
plt.plot(S, short_call_2, linestyle = '--')
plt.title('Short Butterfly Spread')
plt.fill_between(S,short_butterfly,0, where=short_butterfly<0, color='r', alpha=0.1, label = 'Loss')
plt.fill_between(S,short_butterfly,0, where=short_butterfly>0, color='g', alpha=0.1, label = 'Profit')
plt.plot(S, short_butterfly);
```
